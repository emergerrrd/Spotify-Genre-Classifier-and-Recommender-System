### Data Distribution R&B Songs vs. Top 200 R&B Songs

```python
for col in ['danceability', 'energy',\
       'loudness', 'speechiness', 'acousticness', 'instrumentalness',\
       'liveness', 'valence', 'tempo']:
    
#data does not ahve popularity in it
    fig = plt.figure()
    ax= fig.add_subplot(111)
    
    # sns.distplot(df_200[col], ax=ax, label= 'Top 200 R&B')
    # sns.distplot(df_Non200[col], ax=ax, label= 'Non Top 200 R&B')
    # sns.distplot(df2[col], ax=ax, label= 'R&B')
    
    sns.distplot(df_seed_flowg[col], ax=ax, label= 'Flow G', kde_kws={'bw': 0.5}, kde=False)
    sns.distplot(df_seed_kiyo[col], ax=ax, label= 'Kiyo', kde=False)
    sns.distplot(df_seed[col], ax=ax, label= 'Manila Grey', kde=False)
    sns.distplot(df_seed_xxxt[col], ax=ax, label= 'XXXTENTACION', kde=False)
    
#     plt.title("%svs %s: %s " % ('R&B ','Manila Grey',col))
    plt.ylabel('Frequency')
    plt.legend(frameon=False)
    plt.show()
```
